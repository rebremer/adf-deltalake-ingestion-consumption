# Step 0: Create delta lake tables

```
%fs ls dbfs:/mnt/deltamovie/
```

```
%fs ls dbfs:/mnt/deltamovie/2_2_rawdata_delta/_delta_log/
```

```
%sql
DROP TABLE IF EXISTS 2_2_rawdata_delta;
CREATE TABLE 2_2_rawdata_delta
USING DELTA
LOCATION "/mnt/deltamovie/2_2_rawdata_delta"
```

# Step 1: Query the data - raw data, delta

```
%sql

SELECT count(*) FROM 2_2_rawdata_delta
```

```
%sql

SELECT * FROM 2_2_rawdata_delta
ORDER BY version desc
```

```
%sql
DESCRIBE HISTORY 2_2_rawdata_delta
```

```
%sql
-- GET increment
SELECT * FROM 2_2_rawdata_delta VERSION AS OF 2
EXCEPT ALL
SELECT * FROM 2_2_rawdata_delta VERSION AS OF 1
```
